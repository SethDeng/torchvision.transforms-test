```python
from PIL import Image
from torchvision import transforms
import matplotlib.pyplot as plt
```

```python
img = Image.open("test/girl.jpg")
print("original_sizeï¼š",img.size)
img1 = transforms.RandomResizedCrop(224)(img)
print("RandomResizedCrop_size: ",img1.size)

plt.imshow(img),plt.title("original")
plt.show()
plt.imshow(img1),plt.title("RandomResizedCrop")
plt.show()
```

```python
print("original_sizeï¼š",img.size)
img2 = transforms.RandomCrop(224, padding=80)(img)
print("randomcrop_size: ",img1.size)

plt.imshow(img2),plt.title("RandomCrop")
plt.show()
```

```python
print("original_sizeï¼š",img.size)
img3 = transforms.RandomHorizontalFlip()(img)
print("RandomHorizontalFlip_size: ",img3.size)

plt.imshow(img3),plt.title("RandomHorizontalFlip")
plt.show()
```

```python
print("original_sizeï¼š",img.size)
img4 = transforms.ToTensor()(img)
print("ToTensor_size: ",img4.shape)
img5 = transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])(img4)
print("Normalize_size: ",img5.shape)
```
